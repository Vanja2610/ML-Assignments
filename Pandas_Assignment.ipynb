```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

Task 1

```python
dict={0:df['FlightNumber'].iloc[0]}

for i in range(1, df.shape[0]):
    dict[i] = dict[i-1]+10

df['FlightNumber'].fillna(value = dict, inplace=True)
df['FlightNumber'] = df['FlightNumber'].astype('int')
df
```

Task 2

```python
df_temp = df.copy(deep=False)
df_temp
```

```python
df_temp = pd.DataFrame(data=df.values, columns=df.columns)
lst = list(df_temp['From_To'].apply(lambda x: x.split('_')))
df_temp['From'] = [x[0] for x in lst]
df_temp['To'] = [x[1] for x in lst]
df_temp.drop('From_To', axis=1, inplace=True)
df_temp
```

Task 3

```python
for column in ['From', 'To']:
    df_temp[column]= df_temp[column].apply(lambda x: x[0].upper() + x[1:].lower())
    
df_temp
```

Task 4

```python
df.drop('From_To', axis=1,inplace=True)
```

```python
df = pd.concat([df,df_temp], axis=1)
df
```

Task 5

```python
m  = max([len(x) for x in df['RecentDelays'].values])
delays = pd.DataFrame()

for j in range(m):
    column = 'delay_{}'.format(j+1)
    df[column] = 0
    i = 0
    for x in df['RecentDelays'].values:
        try:
            df[column].iloc[i] = x[j] 
        except:
            df[column].iloc[i] = np.nan
        i = i+1
        


df
```
